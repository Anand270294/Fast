# Benchmarking Results for Fast Library
## The results below shows the difference in the computational speed for Fast Library compared to Numpy and a Naive OpenMP C++ implementation (wrapped into Python)

## The was test conducted on a 72 core CPU linux server. 
## The C++ code was compiled used G++ 7 and wrapped using PyBind11.


```python
from matplotlib import pyplot as plt
import time, json
import fast.fast as ft
import numpy as np
```

```python
with open('wrapper_test.json') as json_file:
    data = json.load(json_file)
```

### Integer vaules test : Arrays of integers were used for a element-wise matrix multiplication with sizes from 5 to 5 x 10^9 
### The CPU Type is stated below:

```python
test_type = data['Test']
cpu_used = data['CPU/GPU Used']

print("Test Type : {}".format(test_type))
print("CPU/GPU used: {}".format(cpu_used))
```

```python
x_range = 5* np.logspace(0,9,num=9, endpoint=False, dtype=int)
plt.figure(figsize=(20,10))
plt.subplot(1,2,1)
plt.title('int element-wise multiplication')
plt.xscale('log')
plt.xlabel('Vector Size n')
plt.ylabel('Time Taken (s)')
plt.plot(x_range, data['numpy'],'-o')
plt.plot(x_range, data['cpu'],'-o')
plt.plot(x_range, data['cpu_norm'],'-o')
plt.legend(['Numpy', 'Eigen + MP', 'Naive MP'])
plt.subplot(1,2,2)
plt.xlabel('Vector Size n')
plt.ylabel('log of Time Taken (s)')
plt.xscale('log')
plt.yscale('log')
plt.plot(x_range, data['numpy'],'-o')
plt.plot(x_range, data['cpu'],'-o')
plt.plot(x_range, data['cpu_norm'],'-o')
plt.show()
```
